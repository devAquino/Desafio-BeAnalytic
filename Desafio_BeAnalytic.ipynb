***A validação dos dados deve ocorrer de acordo com a regra de negócio, mas como é um case técnico, criarei uma condição de validação***

***Utilizando regex para criar máscaras para o tratamento das colunas email, phone e postalZip***

```python
mascara_email = '^[a-z]+[\._]?[a-z0-9]+[@]\w+[.](com|edu|net|org|br)'
mascara_phone = '^[(][1-9]{2,2}[)]\s?[0-9]{5,5}-[0-9]{4,4}$'
mascara_postalZip = '^\d{5}-\d{3}$'
```

***Armazenando os valores das colunas em uma nova coluna para garantir a segurançã dos dados caso haja falhas.***

```python
dados_postalZip = dados['postalZip']
dados_phone = dados['phone']
dados_email = dados['email']
```

```python
# O postalZip deve conter 5 valores numérios um hífen e mais 3 valores numéricos.
# Se os dados nesse campo não passarem pela validação, os valores serão substituidos por 00000-000.

#Lista para armazenar os postalZips validados
postalZip_validado = []

def valida_postalZip():
  for postal_zip in dados_postalZip:
    postalZip_valido = re.search(mascara_postalZip, postal_zip)
    if postalZip_valido:
      postalZip_validado.append(postal_zip)
    else:
      postalZip_validado.append("00000-000")
  

valida_postalZip()
```

***Testando a validação do postalZip.***

```python
postal_zip = '23469-123'

if(re.search(mascara_postalZip, postal_zip)):
    print('PostalZip certo')
else:
    print('PostalZip errado')
```

```python
dados_postalZip = postalZip_validado
```

```python
# O phone deve ter 2 números entre 1 e 9 entre parentes, representando o DDD, poderá ter espaço ou não, após o parênteses,
# 5 números entre 0 e 9 seguido de um hífen e mais 4 números entre 0 e 9.

#Lista para armazenar os telefones validados
phone_validado = []

def valida_phone():
  for phone in dados_phone:
    phone_valido = re.search(mascara_phone, phone)
    if phone_valido:
      phone_validado.append(phone)
    else:
      phone_validado.append("(00) 00000-0000")
  return phone_validado

valida_phone()
```

***Testando validação do phone.***

```python
phone = '(11) 22222-1234'

if(re.search(mascara_phone, phone)):
    print('Phone certo')
else:
    print('Phone errado')
```

```python
dados_phone = phone_validado
```

```python
# O email deve iniciar com letra minúscula, pode ter ponto final ou underscore,
# deve conter o arroba, após o arroba deve conter letras, após as letras o ponto seguido por um das
# seguintes opções: (com, edu, net, org, br) 

#Lista para armazenar os emails validados.
email_validado = []

# Função responsável pela validação dos emails.
def valida_email():
  for email in dados_email:
    email_valido = re.search(mascara_email, email)
    if email_valido:
      email_validado.append(email)
    else:
      email_validado.append("invalid email")
  
valida_email()
```

*Testando validação do email.*

```python
email = 'a.f@hotmail.com'

if(re.search(mascara_email, email)):
    print('Endereço certo')
else:
    print('Endereço errado')
```

```python
dados_email = email_validado
```

***Transformando o campo currency em um float.***

```python
dados['currency'] = dados.currency.replace('[^\d.]','', regex=True).astype(float)
```

***Podemos observar que o $ foi removido e que os dados agora são do tipo float.***

```python
dados.currency.head()
```

***Atribuindo os dados validados aos campos originais do dataframe.***

```python
dados['postalZip'] = dados_postalZip
dados['email'] = dados_email
dados['phone'] = dados_phone
```

***Agora podemos observar quais paises tem moedas (currency), com valor menor ou igual a 5.***
